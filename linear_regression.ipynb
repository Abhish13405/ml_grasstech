```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Step 1: DataFrame banaya
# import pandas as pd

df = pd.DataFrame(
    {
        "height": [1, 2, 3, 4, 5, 6, 7, 8, 9],
        "weight": [100, 200, 300, 400, 500, 600, 700, 800, 900],
    }
)


# Step 2: X (input) aur y (output) define karo
x = df["height"]  # double brackets -> DataFrame
y = df["weight"]  # single bracket -> Series

# Step 3: Train aur test data mein split karo
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)

# Step 4: Model banao aur train karo
model = LinearRegression()
model.fit(x_train, y_train)

# Step 5: Prediction karo
y_pred = model.predict([[5]])

# Step 6: Model ka performance check karo
print("R2 Score:", r2_score(y_test, y_pred))
# print("Coefficient:", model.coef_)
# print("Intercept:", model.intercept_)
```
