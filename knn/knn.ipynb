```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
```

```python
data = {
    "Income": [30, 35, 40, 70, 75, 80],
    "SpendingScore": [25, 30, 45, 80, 85, 90],
    "Segment": [
        "regular",
        "regular",
        "regular",
        "Premium",
        "Premium",
        "Premium"
    ]
}
```

```python
df=pd.DataFrame(data)
```

```python
# df
```

```python
# scatter plot
# Budget Customers
regular = df[df["Segment"] == "regular"]

# Premium Customers
premium = df[df["Segment"] == "Premium"]

plt.figure(figsize=(8,6))

# Budget
plt.scatter(
    regular["Income"],
    regular["SpendingScore"],
    color="blue",
    s=100,
    label="regular"
)

# Premium
plt.scatter(
    premium["Income"],
    premium["SpendingScore"],
    color="red",
    s=100,
    label="Premium"
)
for i in range(len(df)):
    plt.text(df["Income"][i]+0.5,
             df["SpendingScore"][i]+0.5,
             f"C{i+1}")
plt.xlabel("Income")
plt.ylabel("Spending Score")
plt.title("Customer Segments")
plt.legend()
plt.grid(True)

plt.show()
```

```python

# Features (X)
X = df[["Income", "SpendingScore"]]

# Target (y)
y = df["Segment"]



# Create KNN Model
model = KNeighborsClassifier(n_neighbors=3,weights="uniform")

```

```python
# Train Model
model.fit(X , y)
```
