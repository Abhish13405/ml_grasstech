```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
# dataset
data={
    "weight":[1,2,3,4,5,6,7,8],
    "sweetness":[6,7,8,9,10,8,9,10],
    "freshness":[8,9,8,10,9,10,9,10],
    "price":[120,250,360,500,620,700,820,950]
}
df=pd.DataFrame(data)
x=df[["weight","sweetness","freshness"]]
y=df["price"]
# Step 3: Train aur test data mein split karo
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
# Step 4: Model banao aur train karo
model = LinearRegression()
model.fit(X_train, y_train)
# Step 5: Prediction karo
y_pred = model.predict(X_test)
print("pridicted price",y_pred)

# accuracy
print("r2_score",r2_score(y_test,y_pred))
new_fruit=[[4,9,9]]
price=model.predict(new_fruit)
print(price)
```
